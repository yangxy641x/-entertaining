# 问题描述

**说有一群兔子，每天回家都会一个一个的经过一个狼窝。狼窝里面的大灰狼有50%的机会抓住经过的兔子。大灰狼抓住3只兔子之后，就不再抓了。问：第六只兔子经过时，被抓的概率有多大? **

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
captured_prob = 0.5
stop_number = 3
rabbits_caught = 0
rabbit_we_care = 5          #python有 0 index，所以输入的数为我们感兴趣的（序数-1），如第六只兔子，rabbit_we_care = 5
def catch_rabbit(captured_prob = 0.5, stop_number = 3, rabbits_caught = 0):
    k = []  #记录兔子捕捉情况
    while True:
        catch_or_not = np.random.choice([1, 0], p=[captured_prob, 1-captured_prob]) #做一次捕捉第k只兔子的行动 
        k.append(catch_or_not)
        rabbits_caught += catch_or_not
        if rabbits_caught == stop_number:
            break
    if len(k) <= rabbit_we_care:
        return 0
    else:
        return k[rabbit_we_care]
```

```python
def sim():
    cir = 1000
    capted = 0
    for i in range(cir):
        a = catch_rabbit(captured_prob, stop_number, rabbits_caught)
        capted += a   
    prob = capted * 100 / cir
    return prob
```
